```python
data_folder = 'data'
#os.makedirs(data_folder, exist_ok=True)

# Save the train_set_copy and test_set_copy DataFrames as CSV files inside the 'data' folder
#train_set_copy.to_csv(os.path.join(data_folder, 'train_data_RNN.csv'), index=False)
#test_set_copy.to_csv(os.path.join(data_folder, 'test_data_RNN.csv'), index=False)

# Read the CSV files from the 'data' folder
train_data = pd.read_csv(os.path.join(data_folder, 'train_data_RNN.csv'))
test_data = pd.read_csv(os.path.join(data_folder, 'test_data_RNN.csv'))
```

```python
train_data.head(3)
```

```python
# Reshaping the training and testing data and scaling it using min_max_scaler
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
trainX_df1 = train_data.iloc[:, :-1]
trainY_df1 = train_data.iloc[:, -1]
print(trainX_df1.shape)
print(trainY_df1.shape)
# Convert trainX_df and trainY_df back to arrays
re_trainX1 = scaler.fit_transform(trainX_df1.to_numpy())
print(re_trainX1.shape)
scaled_trainY1 = scaler.fit_transform(trainY_df1.to_numpy().reshape(-1, 1))
scaled_trainX1 = re_trainX1.reshape(re_trainX1.shape[0], 3, 4)
```

```python
print(scaled_trainX1.shape,scaled_trainY1.shape)
```

```python
#Dividing the training set into train and validation set
final_train_x, val_train_x,final_train_y, val_train_y = train_test_split(scaled_trainX1, scaled_trainY1, test_size=0.1, random_state=42)

```

```python
final_train_x.shape
```

```python
# Training the model using LSTM


from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers

model = Sequential([layers.Input((3, 4)),
                    layers.LSTM(64),
                    layers.Dense(128, activation='relu'),
                    layers.Dense(128, activation='relu'),
                    layers.Dense(1)])

model.compile(loss='mse', 
              optimizer=Adam(learning_rate=0.001),
              metrics=['mean_absolute_error'])
history_callback = model.fit(final_train_x, final_train_y, validation_data=(val_train_x, val_train_y), epochs=1000)

#model.fit(final_train_x, final_train_y, validation_data=(val_train_x, val_train_y), epochs=1000)
```

```python
# TRAINING PREDICTIONS

train_predictions = model.predict(final_train_x).flatten()

plt.plot( train_predictions)
plt.plot( final_train_y)
plt.title('Scaled_predictions')
plt.legend(['Training Predictions', 'Training Observations'])
```

```python
train_pred = train_predictions.reshape(-1,1)
```

```python
train_pred.shape
```

```python
import matplotlib.pyplot as plt
train_predictions_unscaled = scaler.inverse_transform(train_pred.reshape(-1,1))
final_train_y_unscaled = scaler.inverse_transform(final_train_y)



# Assuming you already have 'train_predictions' and 'final_train_y' as arrays containing the training predictions and observations, respectively.

# Create a scatter plot for training predictions and observations
plt.plot(range(len(train_predictions)), train_predictions_unscaled, label='Training Predictions', marker='o')
plt.plot(range(len(final_train_y)), final_train_y_unscaled, label='Training Observations', marker='x')

# Add legend and labels
plt.legend()
plt.xlabel('Data Point Index')
plt.ylabel('Values')
plt.title("Unscaled_Original_Predictions")

plt.show()

```
